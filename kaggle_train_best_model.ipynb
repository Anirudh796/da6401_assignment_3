```python
import os
import shutil

# Optional: disable wandb networking for clean runs
os.environ.setdefault("WANDB_MODE", "disabled")

KAGGLE_INPUT_REPO = "/kaggle/input/datasets/anirudhcs25m010/dl-assignment3-v1/da6401_assignment_3-main"
WORK_REPO = "/kaggle/working/da6401_assignment_3-main"

if not os.path.exists(os.path.join(KAGGLE_INPUT_REPO, "model.py")):
    raise FileNotFoundError(f"Repo not found at {KAGGLE_INPUT_REPO}")

# /kaggle/input is read-only. Copy once to writable /kaggle/working.
if not os.path.exists(WORK_REPO):
    shutil.copytree(KAGGLE_INPUT_REPO, WORK_REPO)

print("Using repo dir:", WORK_REPO)
```

```python
# Full training run via train.py CLI (best checkpoint saved to checkpoint.pt)
!cd /kaggle/working/da6401_assignment_3-main && \
python train.py \
  --mode train \
  --run_name baseline \
  --epochs 30 \
  --batch_size 128 \
  --d_model 256 \
  --N 4 \
  --num_heads 8 \
  --d_ff 1024 \
  --dropout 0.1 \
  --warmup_steps 4000 \
  --learning_rate 1.0 \
  --label_smoothing 0.1 \
  --max_len 100 \
  --patience 6 \
  --checkpoint_path checkpoint.pt \
  --vocab_path vocab.pt \
  --save_best
```
